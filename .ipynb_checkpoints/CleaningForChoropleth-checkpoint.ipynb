```python
import pandas as pd
```

```python
dpi = pd.read_csv('Disposable_Personal_Income.csv')
dpi.head()
```

```python
pce = pd.read_csv('PCE_All_Areas.csv')
pce.head()
```

```python
dpi_state = dpi.loc[dpi["GeoName"] != "United States*",:]
dpi_state = dpi.loc[dpi["TimePeriod"] == 2016,:]
dpi_state.head()
```

```python
pce_state = pce.loc[pce["GeoName"] != "United States",:]
pce_state = pce_state.loc[pce['Description'] == 'Personal consumption expenditures']
pce_state.head()
```

```python
combined_df = pd.merge(dpi_state, pce_state, on="GeoName")
combined_df
```

```python
combined_df.to_csv('data.csv')
```
